```python
import pandas as pd
import numpy as np

# Sample sales data
data = {
    'user_id': [101, 101, 102, 103, 103, 103],
    'item_id': [201, 202, 201, 202, 203, 204],
    'quantity': [1, 2, 1, 1, 2, 3]
}
df = pd.DataFrame(data)

# Function to assign ratings based on quantity
def assign_rating(quantity):
    if quantity == 1:
        return np.random.choice([1, 2, 3])
    else:
        return np.random.choice([4, 5])

# Apply the function to the 'quantity' column
df['rating'] = df['quantity'].apply(assign_rating)

print(df)
```
